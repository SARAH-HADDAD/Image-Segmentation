# Image Segmentation


### We have two object and we're going to try to separte them as much as possible

```python
import cv2 
from matplotlib import pyplot as plt
```

```python
#import the image
img = cv2.imread("img2.jpeg")
# convert our image to RGB 
# all of our colors would be completely mixed up 
img_color=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#convert the image to grayscale
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.figure("Original")
plt.imshow(img_color)
```

```python
plt.figure("Grayscale")
plt.imshow(gray,cmap='gray')
```

###### in this code we are gonna try Otsu Algorithm and triangle algorithm on our image 
###### all binaary segmentation is going to have some threshold value
###### if our pixel is greater than this threshold value then we are going to make that pixel sith some bat max value 
###### we're going to push it all the way to one spectrum 
###### openCV allow us to do that 
###### the return value that OpenCV chooses to be the best candidate to split our image into a foreground and a background
###### then our thresshold is going to hold our image or fresh 

```python
ret, thresh = cv2.threshold(gray,181,255,cv2.THRESH_BINARY)
# only things greater than 128 are going to be pushed 
# every value that's greater than zero is going to be pushed to 255
#if it's not black it's white 
```

```python
plt.imshow(thresh,cmap='gray')
```

```python
#check out Otsu calculation
ret, thresh = cv2.threshold(gray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
# it's gonna igonore the values (0,255) and try to figure out itself
# it's gonna try to determine signle-handedly what the best threshold value is to split something into the foregoround and it's the background  
# maximize the inter class variance 
# we have a brunch of pixels that are sort of close thoghter, it try to push groups all towards zero or 255 
# and it's the potimal way to push the most pixels in the opposite directions 
```

```python
plt.imshow(thresh,cmap='gray')
```

```python
# triangle method 
# the pixel 255 is part of the object 
ret, thresh = cv2.threshold(gray,0,255,cv2.THRESH_TRIANGLE)
print(ret)
```

```python
plt.figure("triangle")
plt.imshow(thresh,cmap='gray')
```
